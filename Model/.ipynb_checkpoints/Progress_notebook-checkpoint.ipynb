# Basic stats

Let's just look at average thicknesses of each of the formations.

```python
keys = df.keys()[5:]
```

```python
import numpy as np

last = np.zeros(len(df))
tdum = []
for key in keys:
    data = df[key]
    x = np.zeros(len(data))
    for i in range(len(data)):
        if type(data[i]) == type(1.2) or type(data[i]) == type(1):
            x[i] = data[i]
        else:
            x[i] = -999.
    total = len(x)
    reported = len(x[x>0])
    thick = x[x>0] - last[x>0]
    print(key , "is present in ", str(round(reported/total * 100.,2)), "% of bores and has an average thickness of " , str(np.average(thick)),"m")
    tdum.append(np.round(np.average(thick),0))
    
    last[x>0.] = x[x>0.]
print(tdum)
print(x)

```

```python
w_all = []
for i in range(len(data)):
    w = []
    for key in keys:
        if type(df.iloc[i][key]) == type(1.) or type(df.iloc[i][key]) == type(1):
            w.append(df.iloc[i][key])
        else:
            w.append(-999.)
    w = np.array(w)
    w_all.append(w)
        
feat = ["TQ_Strat",
       "Tertiary",
       "Tertiary",
       "Tertiary",
       "Tertiary",
       "Coolyena",
       "Coolyena",
       "Coolyena",
       "Coolyena",
       "Coolyena",
       "Coolyena",
       "Coolyena",
       "Warnbro",
       "Warnbro",
       "Warnbro",
       "Warnbro",
       "Warnbro",
       "LKEJ",
       "LKEJ",
       "LKEJ",
       "LKEJ",
       "LKEJ"]   
```
